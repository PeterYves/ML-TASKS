### Problem 1: Number of grains of rice on the 100th day

```python
import numpy as np
arr =[1]
total_num_days = 100
for i in range(1,total_num_days):
    arr.append((arr[i-1])*2)
arr=np.array(arr)
num_grains=np.sum(arr)
print("Total number of rice grains = {}".format(num_grains))
```

- Graph representation

```python

import matplotlib.pyplot as plt

num_days = 101
days = range(1, num_days)

sum_arr=[]
for i in range(100):
    sum_arr.append(np.sum(arr[:i+1]))
print("Sum of rice in", i, "days is: ", num_grains)    

plt.plot(days, arr, 'g-o', linewidth=3, markersize=5, markeredgewidth=2.5, markeredgecolor='orange')
plt.plot(days, sum_arr, 'b^', linewidth=2.5, markersize=5);
plt.title('Number of Rice grains')
plt.xlabel('Days'); 
plt.ylabel('Number of Rice grains');
plt.legend(['Number of rice grains you receive on that day', 'The total number of rice grains you receive by that day'])
plt.show()
```

### Problem 3: How many people can live for how many days with the rice we get?

- according to wikipedia, in a research conducted by WHO it says that " It takes 400grams which is approxmitaly 19200 grains of rice to feed an adult human a day."

```python
#Let's take 200 the number of grains of rice required per one people to live one day
def number_of_days_to_live(number_of_grains,number_of_people):
    
    #by calculating the number of days to live for one people for the given number of grains
    days_to_live_per_one_people = number_of_grains/200
    
    #hence calculate the number of days to live for the given number of people
    num_days_to_live = days_to_live_per_one_people/number_of_people

    return num_days_to_live
n=number_of_days_to_live(19200,5)
print("Number of days to live = {} days".format(n//1))
```
